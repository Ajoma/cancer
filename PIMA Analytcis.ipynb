```python
colnames = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
```

```python
#Create a new column called Nutritional Status

Nutritional_status = pd.Series([])
Nutritional_status
```

```python
#Nutritional status is based on BMI

for i in range(len(df)):
    if df['BMI'][i] == 0.0:
        Nutritional_status[i] = "NA"
    elif df['BMI'][i] < 18.5:
        Nutritional_status[i] = "Underweight"
    elif df['BMI'][i] >= 18.5 and df['BMI'][i] <25:
        Nutritional_status[i] = "Normal"
    elif df['BMI'][i] >= 25 and df['BMI'][i] < 30:
        Nutritional_status[i] = "Overweight"
    elif df['BMI'][i] >= 30:
        Nutritional_status[i] = "Obese"
    else:
        Nutritional_status[i] = df['BMI'][i]
```

```python
#Insert new column Nutritional Status into the existing dataset
df.insert(6, "NutritionalStatus", Nutritional_status)
```

```python
df['NutritionalStatus'].value_counts()
```

11 women BMI data is unavailable. Majority of the women are obese.

```python
# Create an empty series called Glucose_interpretation to interpret the glucose data.
# Then Create a new column called GlucoseInterpretation.
Glucose_interpretation = pd.Series([])
```

```python
for i in range(len(df)):
    if df['Glucose'][i] == 0.0:
        Glucose_interpretation[i] = "NA"
    elif df['Glucose'][i] <=140:
        Glucose_interpretation[i] = "Normal"
    elif df['Glucose'][i] > 140 & df['Glucose'][i] <=198:
        Glucose_interpretation[i] = "Impaired Glucose Tolerance"
    elif df['Glucose'][i] > 198:
        Glucose_interpretation[i] = "Diabetic Level"
    else:
        Glucose_interpretation[i] = df['Glucose'][i]
```

```python
## Create a new column called GlucoseInterpretation.
df.insert(2, 'Glucose Interpretation', Glucose_interpretation)
```

Logic Regression

```python
var1 = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
x=df_all[var1]
y=df_all.Outcome
```

```python
#Define the model and assign dependent variable(y) and independent variable(x)
logit_model =sm.Logit(y,x)
result=logit_model.fit()
print(result.summary())
```
